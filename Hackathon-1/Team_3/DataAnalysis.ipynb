<h3> We first try to understand what each feature means after looking at the data</h3>
<h4>animal_id_outcome</h4> - Is the id given to the animal to identify it.
<h4>dob_year</h4> - Is the year the animal was born
<h4>dob_month</h4> - Is the month the animal was born
<h4>age_upon_intake</h4> - Is the age of animal at the time of intake
<h4>animal_type</h4> - The type of animal ('Dog', 'Cat', 'Other', 'Bird')
<h4>breed</h4> - The breed of the animal
<h4>color</h4> - The color of the animal
<h4>intake_condition</h4> - The intake condition of the animal ('Normal', 'Injured', 'Aged', 'Sick', 'Other', 'Feral', 'Pregnant', 'Nursing')
<h4>intake_type</h4> - The intake type of the animal ('Stray', 'Public Assist', 'Owner Surrender', 'Euthanasia Request', 'Wildlife')
<h4>sex_upon_intake and sex_upon_outcome</h4> - The sex of the animal at the time of intake and outcome ('Neutered Male', 'Spayed Female', 'Intact Female', 'Intact Male', 'Unknown')
<h4>count</h4> - As the value of count is 1 for all the rows, we are not instantly able to understand what count exactly is.
<h4>age_upon_intake_(days), age_upon_intake_(years)</h4> -  The age of the animal during intake in days and years.
<h4>age_upon_intake_age_group</h4> - The age upon intake is divided into groups and made as categorical data. The categories are ('(7.5, 10.0]', '(5.0, 7.5]', '(15.0, 17.5]', '(12.5, 15.0]', '(17.5, 20.0]', '(10.0, 12.5]', '(2.5, 5.0]', '(-0.025, 2.5]', '(20.0, 22.5]')
<h4>intake_datetime, intake_month, intake_year, intake_monthyear, intake_weekday, intake_hour</h4> - The datetime, month, year, monthyear, weekday and hour respectively at the time of intake.
<h4>intake_number</h4>  - The number at intakes
<h4>time_in_shelter, time_in_shelter_days</h4> - The time in shelter in days,time and in days respectively.
<h4>age_upon_outcome, age_upon_outcome_(days), age_upon_outcome_(years)</h4> -  The age of the animal at the outcome in days and years.
<h4>age_upon_outcome_age_group</h4> - The age at outcome is divided into groups and made as categorical data. The categories are ('(7.5, 10.0]', '(5.0, 7.5]', '(15.0, 17.5]', '(12.5, 15.0]', '(17.5, 20.0]', '(10.0, 12.5]', '(2.5, 5.0]', '(-0.025, 2.5]', '(20.0, 22.5]')
<h4>outcome_datetime, outcome_month, outcome_year, outcome_monthyear, outcome_weekday, outcome_hour</h4> - The datetime, month, year, monthyear, weekday, hour upon outcome.
<h4>outcome_number</h4> - The number of animals at outcome. We also realise that number at outcome and intake are always same in this dataset.
<h4>outcome_type</h4> - The type of outcome ('Return to Owner', 'Missing', 'Transfer', 'Adoption', 'Euthanasia', 'Rto-Adopt', 'Relocate', 'Died', 'Disposal')

But we can use other ways such as maximum count of sex of all the animals or the sex of the same type and breed of animal to fill the missing data.<br><br>
We can fill the missing values in outcome_datetime using intake_datetime and time_in_shelter as both of them have zero(0) missing(null) values. outcome_datetime = intake_datetime + time_in_shelter.


<h3>Filling the missing values in 'outcome_datetime' feature</h3>

<h4>The next few cells are to test our method to calculate 'outcome_datetime' by using the formula : outcome_datetime = intake_datetime + time_in_shelter<h4>

```python
outcome_datetime_null_values = shelter_df[shelter_df['outcome_datetime'].isnull()]
print(len(outcome_datetime_null_values))
```

```python
print(outcome_datetime_null_values.loc[2, 'intake_datetime'], outcome_datetime_null_values.loc[2, 'time_in_shelter'], outcome_datetime_null_values.loc[2, 'outcome_datetime'])
```

```python
from datetime import datetime, timedelta
print(shelter_df.loc[0,'outcome_datetime'],  datetime.strptime(shelter_df.loc[0, 'intake_datetime'], '%Y-%m-%d %H:%M:%S') + (timedelta(seconds = shelter_df.loc[0, 'time_in_shelter_days'] * 3600 * 24)))
```

```python
print(outcome_datetime_null_values.loc[2, 'intake_datetime'], outcome_datetime_null_values.loc[2, 'time_in_shelter'], outcome_datetime_null_values.loc[2, 'outcome_datetime'])
print(datetime.strptime(outcome_datetime_null_values.loc[2, 'intake_datetime'], '%Y-%m-%d %H:%M:%S') + (timedelta(seconds = outcome_datetime_null_values.loc[2, 'time_in_shelter_days'] * 3600 * 24)))
```

```python
print(len(shelter_df[shelter_df['outcome_datetime'].isnull()]))
```

<h4>Filling the missing values with -1, so that we can compare it and change with the required value in the next cell</h4

```python
shelter_df['outcome_datetime'].fillna(-1, inplace = True)
```

<h4> Converting string to datetime object, calculating the required datetime and then converting back to string</h4>

```python
for i in range(len(shelter_df)) :
    if(shelter_df.loc[i, 'outcome_datetime'] == -1) :
        shelter_df.loc[i, 'outcome_datetime'] =  (datetime.strptime(shelter_df.loc[i, 'intake_datetime'], '%Y-%m-%d %H:%M:%S') + (timedelta(seconds = shelter_df.loc[i, 'time_in_shelter_days'] * 3600 * 24))).strftime('%Y-%m-%d %H:%M:%S')
```
